```python
import numpy as np
import pylab as pl
import scipy as sp
from scipy import stats
```

```python
np.random.seed(109)
```

```python
fig, ax = pl.subplots()
h = []
for a in [0.2, 0.6, 1.0]: # a = a(T)
  d = stats.chi.rvs(df = 3, scale = a, size = 1000) # distribuzione di Maxwell equivalente al chi con 3 gradi di libertà
  bins = np.arange(0, 5, 0.1)
  h.append(ax.hist(d, # stats.maxwell.rvs(scale = a, size = 1000), 
          bins, alpha = 0.7, label = "a = %.1f"%a));
ax.legend();
ax.set_xlabel('speed');
ax.set_ylabel('N'); # numero di particelle nel range di velocità
ax.set_title('Maxwell distributions');
```

a = a(T)

particelle a T > raggiungono velocità > --> $a = \sqrt T$

```python
# media
df = 3
a = [0.2, 0.6, 1.0]
mean = np.array(a) * np.sqrt(2) * sp.special.gamma((df+1)/2) / sp.special.gamma(df/2) # si moltiplica la media della distribuzione per a
mean
```

```python
# velocità più probabile
Nmax = []
for i in range(3):
  b = h[i][1][:-1] # si considera l'estremo sx dell'intervallo
  Nmax.append(b[h[i][0] == np.max(h[i][0])])
Nmax
```
