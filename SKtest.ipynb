```python
import scipy as sp
from scipy import stats
import pylab as pl
import numpy as np
```

```python
sp.stats.norm.rvs(size = 100) # rvs: random values
```

```python
sp.stats.kstest # sostare col mouse per info
# confronto di una variabile random con una data distribuzione da un modello teorico
# stats.kstest(x, 'norm') --> 'norm' di default, ma si può cambiare
# sotto ipotesi nulla le due distribuzioni sono uguali
```

paragonare distribuzioni normali e la distribuzione normale con un'altra distribuzione

generate N values from  a distribution (normal fist then another one)
Use the spicy.stats.kstest() to test if the sample is normal or not

```python
normDistr = sp.stats.norm.rvs(size = 100)
```

```python
sp.stats.kstest(normDistr, 'norm') # il test dice se la distribuzione in input è una distribuzione normale
# p-value > 0.05 --> non si rigetta l'ipotesi nulla, le due distribuzioni si somigliano
```

I have not rejected that the samples come from the same distribution

```python
binomDistr = sp.stats.binom.rvs(10, 0.5, size = 100)
```

```python
sp.stats.kstest(binomDistr, 'norm')
```

devo rigettare l'ipotesi che la distribuzione venga da quella normale se p < 0.05

```python
binomDistr = sp.stats.binom.rvs(1000, 0.5, size = 100)
```

```python
std = np.std(binomDistr)
mean = np.mean(binomDistr)
pl.hist(binomDistr);
```

```python
binomDistr_std = (binomDistr - mean) / std # confronto con distribuzione a media 0 e std 1 --> bisogna standardizzare la binomiale
```

```python
pl.hist(binomDistr_std);
```

```python
sp.stats.kstest(binomDistr_std, 'norm') # la binomiale si avvicina alla normale e il p value è alto
```
