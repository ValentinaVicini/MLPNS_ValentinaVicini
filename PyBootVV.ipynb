EXERCISE

* create a function that takes a string
* checks if the string is a palindrome
* returns 1 if true
* return -1 if false
* and also prints "Yes its a palindrome" or "this is not a palindrome"
* add a print statement that prints ONLY the unique part (rad in radar)

```
def palindrome(word):
  if not type(word) == str:
    return('Pass a string')
  for c in "',.;: ": # elimina caratteri indesiderati
    word = word.replace(c,'')
  if word.lower() == word.lower()[::-1]: # lower per non avere maiuscole
    print('Yes, its a palindrome')
    print(word[:int(len(word)/2 + 0.5)])
    return 1
  else:
    print('This is not a palindrome')
    return -1
```

```
palindrome("Ai lati d'Italia")
```
