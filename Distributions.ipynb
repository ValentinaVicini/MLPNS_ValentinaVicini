```python
import numpy as np
import scipy as sp
from scipy import stats
import pylab as pl
```

# Distribuzione Gaussiana
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html

```python
stats.norm?
# A normal continuous random variable.
# The location (``loc``) keyword specifies the mean.
# The scale (``scale``) keyword specifies the standard deviation.
# rvs(loc=0, scale=1, size=1)
# pdf(x, loc=0, scale=1) --> probability density function
# cdf(x, loc=0, scale=1) --> cumulative distribution function
```

```python
pl.hist(stats.norm.rvs(0, 1, 100));
```

```python
pl.hist(stats.norm.rvs(100, 10, 100));
```

```python
pl.plot(stats.norm.cdf(np.linspace(-10,10,100))) # plot e non hist!
# considera la distribuzione standard tra -10 e 10 suddivisa in 100 intervalli
```

```python
pl.plot(stats.norm.pdf(np.linspace(-7,7,100))) # plot e non hist!
```

# Distribuzione di Poisson
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html

```python
stats.poisson?
# A Poisson discrete random variable.
# rvs(mu, loc=0, size=1)
# pmf(k, mu, loc=0) --> probability mass function
# cdf(k, mu, loc=0) --> cumulative distribution function
```

```python
pl.hist(stats.poisson.rvs(7, 0, size = 100));
# il primo valore corrisponde sia alla media che alla varianza
# il secondo valore trasla la media --> da 7 a 17
```

```python
pl.plot(stats.poisson.cdf(np.linspace(0,14,1000), 7, 0)); # funzione discreta
```

# Distribuzione binomiale
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.binom.html

```python
stats.binom?
# A binomial discrete random variable.
# rvs(n, p, loc=0, size=1)
# n: numero di tentativi
# p: probabilit di successo per ogni tentativo
# pmf(k, n, p, loc=0) --> probability mass function
# cdf(k, n, p, loc=0) --> cumulative distribution function
```

```python
pl.hist(stats.binom.rvs(100, 0.5, 0, size = 1000));
```

```python
pl.plot(stats.binom.cdf(np.linspace(30,70,1000), 100, 0.5, 0));
```

# Distribuzione del $\chi ^2$
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2.html

```python
stats.chi2?
# A chi-squared continuous random variable.
# rvs(df, loc=0, scale=1, size=1)
# pdf(x, df, loc=0, scale=1)
# cdf(x, df, loc=0, scale=1)
```

```python
for i in [3, 5, 7]:
  pl.hist(stats.chi2.rvs(i, size = 1000), alpha = 0.7, label = '%d'%i);
pl.legend();
```

```python
for i in [3, 5, 7]:
  pl.plot(stats.chi2.cdf(np.linspace(0, 20, 100), i, loc=0, scale=1), label = '%d'%i);
pl.legend();
```
